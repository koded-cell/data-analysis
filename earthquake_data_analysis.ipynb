```python
mean_magnitude = df['Magnitude'].mean()
median_magnitude = df['Magnitude'].median()
print(f"Mean Magnitude: {mean_magnitude}, Median Magnitude: {median_magnitude}")
```

```python
yyearly_data = df.groupby('Year')['Magnitude'].mean()
yearly_data.plot(kind='line')
plt.title('Average Earthquake Magnitude per Year')
plt.xlabel('Year')
plt.ylabel('Average Magnitude')
plt.show()
```

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
```

```python
df['Year'] = pd.to_datetime(df['Date'], format='ISO8601').dt.year
```

```python
df['Year'] = pd.to_datetime(df['Date'], format='%Y-%m-%dT%H:%M:%S.%fZ').dt.year
```

```python
df['Year'] = pd.to_datetime(df['Date'], format='mixed', errors='coerce').dt.year
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', utc=True)
df['Year'] = df['Date'].dt.year
df.head()
```

```python
yearly_data = df.groupby('Year')['Magnitude'].mean()
yearly_data.plot(kind='line')
plt.title('Average Earthquake Magnitude per Year')
plt.xlabel('Year')
plt.ylabel('Average Magnitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Magnitude'], bins=30, kde=True)
plt.title('Distribution of Eaethquake Magnitude')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
yearly_magnitude = df.groupby('Year')['Magnitude'].mean()
plt.figure(figsize=(10, 6))
plt.plot(yearly_magnitude.index, yearly_magnitude.values, marker='o')
plt.title('Trend of Earthquake Magnitudes Over the Years')
plt.xlabel('Year')
plt.ylabel('Mean Magnitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Depth', y='Magnitude', data=df)
plt.title('Relationship Between Earthquake Depth and Magnitude')
plt.xlabel('Depth (Km)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
correlation = df['Depth'].corr(df['Magnitude'])
print(f'Correlation between Depth and Magnitude: {correlation}')
```
